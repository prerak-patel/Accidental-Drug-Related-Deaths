```python
#What are the drugs of choice for this location?

import pandas as pd 
import matplotlib.pyplot as plt 
import scipy.stats as st 
import numpy as np 
import random 

#data file 
accidental_drug_related = "Accidental_Drug_Related_Deaths_2012-2018.csv"

#read csv data and analyze results 
accidental_drug_related = pd.read_csv(accidental_drug_related)
accidental_drug_related.head()
```

```python
accidental_drug_related.drop(['ID','Date','DateType','Age','Sex','Race','MannerofDeath',
'DeathCityGeo','ResidenceCityGeo','InjuryCityGeo','DescriptionofInjury','InjuryPlace','InjuryCity',
'InjuryCounty','InjuryState','COD','OtherSignifican','OpiateNOS','AnyOpioid','MannerofDeath',
'DeathCityGeo','ResidenceCityGeo','InjuryCityGeo','Location','LocationifOther'], axis=1, inplace=True)
```

```python
accidental_drug_related.head()
```

```python
accidental_drug_related.isnull().sum()
```

```python
accidental_drug_related.drop(['ResidenceCity','ResidenceCounty','ResidenceState','DeathCounty'], axis=1, inplace=True)

accidental_drug_related.head()
```

```python
accidental_drug_related.shape

accidental_drug_related.dropna(subset=['DeathCity'], how='any').shape

accidental_drug_related.head(10)
```

```python
#accidental_drug_related Heroin
adr_heroin = accidental_drug_related[accidental_drug_related.Heroin=='Y']

adr_heroin_count = adr_heroin.groupby('DeathCity').Heroin.count()
adr_heroin_impact = adr_heroin_count.nlargest(20)

#create a bar chart showing the information 
plt.xdata = adr_heroin_impact.plot.bar(legend=False)
plt.xlabel('City')
plt.ylabel('Number of Deaths')
plt.title('Number of Deaths by Heroin')

plt.show()
```
