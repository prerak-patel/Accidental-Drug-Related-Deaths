```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
drug_related_deaths_df = pd.read_csv('Accidental_Drug_Related_Deaths_2012-2018 (2).csv')
# drug_related_deaths_df
```

```python
narrowed_df = drug_related_deaths_df.loc[:,['Date','Age','Sex','Race','DeathCity','DeathCityGeo','COD','Heroin','Cocaine','Fentanyl','FentanylAnalogue','Oxycodone','Oxymorphone','Ethanol','Hydrocodone','Benzodiazepine','Methadone','Amphet','Tramad','Morphine_NotHeroin','Hydromorphone','Other','OpiateNOS','AnyOpioid']]
```

```python
# Drop records without Age, Sex and Race entry
narrowed_df = narrowed_df.dropna(subset=['Date', 'Age', 'Sex','Race'], how='all')
```

```python
narrowed_df.update(narrowed_df[['Heroin','Cocaine','Fentanyl','FentanylAnalogue','Oxycodone','Oxymorphone','Ethanol','Hydrocodone','Benzodiazepine','Methadone','Amphet','Tramad','Morphine_NotHeroin','Hydromorphone','Other','OpiateNOS','AnyOpioid']].fillna('N'))
narrowed_df['Date'] = pd.to_datetime(narrowed_df['Date']).dt.strftime('%Y')
narrowed_df
```

```python
narrowed_df['Date'] = narrowed_df['Date'].astype(float)
narrowed_df['Date']
```

```python
# narrowed_df['Date'].value_counts()
narrowed_df['Date'].unique()
```

```python
datapoints = narrowed_df.groupby("Date")['Date'].value_counts()

#Do a linear regression 

# Use DataFrame.plot() in order to create a bar chart of the data
datapoints.plot(kind="bar", color = "r", legend=False)

# Set a title for the chart
plt.title("Number of Accidental Deaths Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Deaths")

plt.show()
plt.tight_layout()
```

```python
cocainepoints = narrowed_df.groupby("Date")['Heroin'].value_counts()
# Use DataFrame.plot() in order to create a bar chart of the data
cocainepoints.plot(kind="bar", color = "r", legend=False)

# Set a title for the chart
plt.title("Heroin Use Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Deaths")

plt.show()
plt.tight_layout()
```
